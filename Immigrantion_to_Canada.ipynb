# The Dataset: Immigration to Canada from 1980 to 2013 

The dataset contains annual data on the flows of international immigrants as recorded by the countries of destination. The data presents both inflows and outflows according to the place of birth, citizenship or place of previous / next residence both for foreigners and nationals. 


```python
import numpy as np
import pandas as pd
```

Read data from excel file using read_excel function

```python
df = pd.read_excel('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.xlsx',
                        sheet_name='Canada by Citizenship',
                       skiprows=range(20),
                       skipfooter=2)

print ('Data read into a pandas dataframe!')
```

```python
df.head()
```

```python
df.columns.values
```

Similarly, to get the list of indicies we use the .index parameter.

```python
df.index.values
```

```python
print(type(df.columns))
print(type(df.index))
```

To get the index and columns as lists, we can use the tolist() method.

```python
print(df.columns.tolist())
print(df.index.tolist())
```

# Data Cleanining

Delete all the columns that are unnecessary using drop() method

```python
df.drop(['AREA', 'REG', 'DEV', 'Type', 'Coverage'], axis = 1, inplace = True)
df.head()
```

Let's rename the columns so that they make sense. Use rename() method by passing in a dictionary of old and new names as follows:

```python
df.rename(columns = {'OdName':'Country', 'AreaName':'Continent', 'RegName': 'Region'}, inplace = True)
```

```python
df.head()
```

 Lets add a 'Total' column that sums up the total immigrants by country over the entire period 1980 - 2013, as follows

```python
df['Total']= df.sum(axis=1)
```

```python
df.head()
```

check to see how many null objects we have in the dataset as follows:

```python
df.isnull().sum()
```

Before we proceed, notice that the defaul index of the dataset is a numeric range from 0 to 194. This makes it very difficult to do a query by a specific country. For example to search for data on Japan, we need to know the corressponding index value.

This can be fixed very easily by setting the 'Country' column as the index using set_index() method.

```python
df.set_index('Country', inplace=True)
df.head()
```

Sort the column Total so that we can see the 

```python
df.sort_values(by=['Total'], ascending=False,inplace = True)
```

View top 5 country with highest immigrants

```python
df_top5 = df.head(5)
df_top5
```

# Visualizing Data using Matplotlib

```python
# we are using the inline backend
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt
```

Column names that are integers (such as the years) might introduce some confusion. For example, when we are referencing the year 2013, one might confuse that when the 2013th positional index.

To avoid this ambuigity, let's convert the column names into strings: '1980' to '2013'.

```python
df.columns = list(map(str, df.columns))
# [print (type(x)) for x in df_can.columns.values] #<-- uncomment to check type of column headers
```

Since we converted the years to string, let's declare a variable that will allow us to easily call upon the full range of years:

```python
years = list(map(str, range(1980, 2014)))
years
```

Visualize the immigrants from Germany

```python
Germany = df.loc['Germany', years] # passing in years 1980 - 2013 to exclude the 'total' column
```

```python
Germany.head()
```

```python
Germany.plot(kind='line')

plt.title('Immigration from Germany')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show() 
```

Visualize immigrants from India 

```python
India = df.loc['India', years]
```

```python
India.plot(kind='line')

plt.title('Immigration from India')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show() 
```

Let's compare the number of immigrants from India and China from 1980 to 2013.

```python
df_CI = df.loc[['India', 'China'], years]
df_CI.head()
```

Plot graph. We will explicitly specify line plot by passing in kind parameter to plot()

```python
df_CI.plot(kind='line')
```

That doesn't look right...

Recall that pandas plots the indices on the x-axis and the columns as individual lines on the y-axis. Since df_CI is a dataframe with the country as the index and years as the columns, we must first transpose the dataframe using transpose() method to swap the row and columns.

```python
df_CI = df_CI.transpose()
df_CI.head()
```

```python
df_CI.index = df_CI.index.map(int) # let's change the index values of df_CI to type integer for plotting
df_CI.plot(kind='line')

plt.title('Immigrants from China and India')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')

plt.show()
```

Compare the trend of top 5 countries that contributed the most to immigration to Canada.

```python
# Take top5 countries into new dataframe df_top5 
df_top5 = df.head(5)

# transpose the dataframe  
df_top5 = df_top5[years].transpose() 

print(df_top5)



df_top5.index = df_top5.index.map(int) # let's change the index values of df_top5 to type integer for plotting
df_top5.plot(kind='line', figsize=(14, 8)) # pass a tuple (x, y) size



plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')


plt.show()
```
